Stacked bar plot for Region Sales group by “Genre” attribute.

```
# import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import RandomState

df_csv = pd.read_csv('data.csv')

Genre = df_csv['Genre']
NA_Sales = df_csv['NA_Sales']
EU_Sales = df_csv['EU_Sales']
JP_Sales = df_csv['JP_Sales']
ind = [country for country in Genre]
 
plt.bar(ind, JP_Sales, width=0.6, label='Japan', color='green', bottom=EU_Sales+NA_Sales)
plt.bar(ind, EU_Sales, width=0.6, label='Eu', color='blue', bottom=NA_Sales)
plt.bar(ind, NA_Sales, width=0.6, label='Na', color='red')
 
plt.xticks(ind, Genre)
plt.ylabel("Sales")
plt.xlabel("Genres")
plt.legend(loc="upper right")
plt.title("Sales by genre and region")
plt.show()
```
